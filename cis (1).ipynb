```python
def format_duration(iso_duration):
    """ISO 8601å½¢å¼ã®å†ç”Ÿæ™‚é–“ã‚’æ™‚:åˆ†:ç§’å½¢å¼ã«å¤‰æ›"""
    duration = isodate.parse_duration(iso_duration)
    return str(duration)
```

```python
def search_youtube(api_key, query, max_results=10):
    """ã‚¯ã‚¨ãƒªã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ã‹ã‚‰æ¤œç´¢çµæœã‚’å–å¾—"""
    url = "https://www.googleapis.com/youtube/v3/search"
    params = {
        "part": "snippet",
        "q": query,
        "maxResults": max_results,
        "type": "video",
        "key": api_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        return None
```

```python
def get_video_details(api_key, video_id):
    """å‹•ç”»IDã‹ã‚‰ã‚¿ã‚¤ãƒˆãƒ«ã¨è¦–è´å›æ•°ã€é«˜è©•ä¾¡æ•°ã€ã‚³ãƒ¡ãƒ³ãƒˆæ•°ã€æŠ•ç¨¿å¹´æœˆæ—¥ã€å‹•ç”»æ™‚é–“ã‚’å–å¾—"""
    url = "https://www.googleapis.com/youtube/v3/videos"
    params = {
        "part": "snippet,statistics, contentDetails",
        "id": video_id,
        "key": api_key
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if "items" in data and len(data["items"]) > 0:
            video_info = data["items"][0]
            title = video_info["snippet"]["title"]

            view_count = int(video_info["statistics"]["viewCount"])

            try:
                like_count = int(video_info["statistics"]["likeCount"])
            except KeyError:
                like_count = "éå…¬é–‹"

            try:
                comment_count = int(video_info["statistics"]["commentCount"])
            except KeyError:
                comment_count = "éå…¬é–‹"
                
            duration = video_info["contentDetails"]["duration"]

            duration_seconds = int(isodate.parse_duration(duration).total_seconds()) #å‹•ç”»æ™‚é–“ã‚’ç§’æ•°ã«å¤‰æ›

            utc_published_at = data['items'][0]['snippet']['publishedAt'] #æŠ•ç¨¿å¹´æœˆæ—¥å–å¾—

            utc_dt = parser.isoparse(utc_published_at)

            jst = pytz.timezone('Asia/Tokyo')
            jst_dt = utc_dt.astimezone(jst)


            return {
                "title": title,
                "view_count": view_count,
                "like_count": like_count,
                "comment_count": comment_count,
                "duration": duration,
                "duration_seconds": duration_seconds,
                "jst_dt" : jst_dt.strftime('%Y-%m-%d')
                }
        
    return None
```
